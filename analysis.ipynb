# U.S. Medical Insurance Costs

Goal: Investigate the Impact of Smoking on Insurance Charges
Analytical Steps:
* Compare the distribution of insurance charges for smokers and non-smokers using descriptive statistics and visualizations.
* Calculate the average difference in charges between smokers and non-smokers.
* Conduct a hypothesis test to determine if the difference in charges is statistically significant.
* Consider controlling for other factors such as age, BMI, and number of children to isolate the effect of smoking on charges.

```python
import csv
insurance_member = []
with open("insurance.csv") as insurance_csv:
    user_reader = csv.DictReader(insurance_csv)
    for row in user_reader:
        insurance_member.append(row)
smoker = []
non_smoker = []
for smoke_status in insurance_member:
    if smoke_status['smoker'] == 'yes':
        smoker.append(smoke_status['charges'])
    else:
        non_smoker.append(smoke_status['charges'])

print(smoker)
print(non_smoker)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Convert charges from strings to floats
smoker_charges = [float(charge) for charge in smoker]
non_smoker_charges = [float(charge) for charge in non_smoker]

#Calculate Descriptive Anaylsis
smoker_mean = np.mean(smoker_charges)
smoker_median = np.median(smoker_charges)
smoker_std = np.std(smoker_charges)

non_smoker_mean = np.mean(non_smoker_charges)
non_smoker_median = np.median(non_smoker_charges)
non_smoker_std = np.std(non_smoker_charges)
# Plot the distributions
plt.hist(smoker_charges, bins=30, alpha=0.5, label='Smoker')
plt.hist(non_smoker_charges, bins=30, alpha=0.5, label='Non-Smoker')
plt.xlabel('Insurance Charges')
plt.ylabel('Frequency')
plt.title('Distribution of Charges: Smoker vs Non-Smoker')
plt.legend()
# Add the Descriptive Anaylsis

plt.text(0.75, 0.95,f"Smoker: mean = {smoker_mean:.2f}, median = {smoker_median:.2f}, std = {smoker_std:.2f}", transform = plt.gca().transAxes )
plt.text(0.75)
# Show the plot
plt.show()

```
